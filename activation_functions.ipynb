```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
x = torch.linspace(-5, 5, 10)
x
```

```python
sigmoid = torch.sigmoid(x)
sigmoid
```

```python
tanh = torch.tanh(x)
tanh
```

```python
relu = F.relu(x)
relu
```

```python
alpha: float = 0.1
leakyRelu = F.leaky_relu(x, negative_slope = alpha)
leakyRelu
```
