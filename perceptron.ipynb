### Setup

```python
import torch
import torch.nn as nn
```

### Perceptron implementation

```python
class Perceptron(nn.Module):

     def __init__(self, inputDimension):
          super().__init__()
          self.linearLayer = nn.Linear(inputDimension, 1)

     def forward(self, x):
          return torch.sigmoid(self.linearLayer(x))
```

```python
model = Perceptron(inputDimension = 4)
```

```python
model
```

```python
model(torch.tensor([
     1.0, 
     2.0, 
     3.0, 
     4.0
]))
```

```python
model(torch.tensor([
     [1.0, 2.0, 3.0, 4.0],
     [2.5, 1.5, 4.0, 1.2]
]))
```
