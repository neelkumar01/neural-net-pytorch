### Setup

```python
import torch
import torch.nn as nn
```

### MLP implementation

```python
class MLP(nn.Module):

     def __init__(
               self,
               inputDim: int,
               hiddenDim: int,
               outputDim: int
     ):
          super().__init__()
          self.layers = nn.Sequential(
               nn.Linear(inputDim, hiddenDim),
               nn.ReLU(),
               nn.Linear(hiddenDim, outputDim)
          )
     
     def forward(self, x):
          return self.layers(x)
```

```python
model = MLP(
     inputDim=3,
     hiddenDim=5,
     outputDim=2
)
```

```python
model
```

```python
x = torch.randn(10, 3)
output = model(x)
```
