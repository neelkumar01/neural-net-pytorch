```python
import torch
```

```python
x = torch.tensor([2.0, 3.0], requires_grad=True)
w = torch.tensor([1.5, -0.5], requires_grad=True)
b = torch.tensor([1.0], requires_grad=True)
```

```python
activation = x @ w + b
activation
```

```python
output = torch.sigmoid(activation)
output
```

```python
target = torch.tensor([1.0])
target
```

```python
loss = (output - target) ** 2
loss
```

```python
loss.backward()
```

```python
x.grad
```
